# Matplotlib

```python
# Hacer que los resultados de los graficos se muestren en el cuaderno
%matplotlib inline
```

```python
# Importar la libreria de dibujo
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4]
```

```python
plt.plot(x)
plt.xlabel("Eje de abcisas")
plt.ylabel("Eje de ordenadas")
plt.show()
```

```python
x = [1,2,3,4]
y = [1,4,9,16]
plt.plot(x,y)
```

```python
# La primera letras es el color y la segunda es el tipo 
plt.plot(x,y, "ro")
# Asignamos que parte del cuadrante vemos
plt.axis([0,6,0,20])
plt.show()
```

```python
import numpy as np
```

```python
data = np.arange(0.0,10.0, 0.2)
data
```

```python
plt.plot(data, data, "r--", data, data**2, "bs", data, data**3, "g^")
plt.show()
```
