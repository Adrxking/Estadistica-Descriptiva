## Trabajar con multiples graficos en una misma figura

```python
def f(x):
    return np.exp(-x)*np.cos(2*np.pi*x)
```

```python
x1 = np.arange(0,5.0,0.1)
x2 = np.arange(0,5.0,0.2)
```

```python
# Definir la figura
plt.figure(1)
# 211 quiere decir doble de ancho que de alto. El ultimo numero es el numero de la figura
plt.subplot(211)
plt.plot(x1, f(x1), 'ro', x2, f(x2), 'k')
plt.subplot(212)
plt.plot(x2, f(x2), 'g--')
plt.show()
```

```python
plt.figure(1)
plt.subplot(2,1,1)
plt.plot([1,2,3])
plt.subplot(2,1,2)
plt.plot([4,5,6])

plt.figure(2)
plt.plot([4,5,6])

plt.figure(1)
plt.subplot(2,1,1)
plt.title("Esto es el primer titulo")
plt.subplot(2,1,2)
plt.show()
```

## Texto en los graficos
(Histograma de frecuencias)

```python
# media
mu = 100
# desviacion tipica
sigma = 20

x = mu + sigma * np.random.randn(10000)
```

```python
# Histograma de frecuencias
## bins son las divisiones
## patches
n, bins, patches = plt.hist(x, 50, facecolor="g", alpha=0.6)

plt.xlabel("Cociente Intelectual", fontsize = 14, color = "green")
plt.ylabel("Probabilidad")
plt.title(r"Histograma de CI $N(\mu, \sigma)$")
plt.text(30,400, r'$\mu = 100, \ \sigma=20$')
plt.axis([0,180,0, 700])
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(10,6), dpi=90)
plt.subplot(111)

x = np.arange(0,10*np.pi, 0.01)

y = np.cos(x)

plt.plot(x,y, lw = 2.0)

# Anotar texto
plt.annotate("Mximo local", xy=(4*np.pi, 1), xytext=(15,1.5),
            arrowprops=dict(facecolor="black", shrink=0.05))

# Expandir eje de las y
plt.ylim(-2,2)

plt.show()
```

## Cambios en los ejes 

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
S, C = np.sin(x), np.cos(x)
```

```python
plt.figure(figsize=(10,8))

plt.plot(x,S, color = "blue", linewidth = 1.2, linestyle = "-", label = "seno")
plt.plot(x,C, color = "green", linewidth = 1.2, linestyle = "-", label = "coseno")
# Aumentar en 10% el tamao del lienzo
plt.xlim(x.min()*1.1, x.max()*1.1)
plt.ylim(S.min()*1.1, S.max()*1.1)

plt.xticks([-np.pi, -np.pi/2.0, 0, np.pi/2, np.pi],
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'+$\pi/2$', r'+$\pi$'])
plt.yticks(np.linspace(-1,1, 2, endpoint=True),
          ['-1', '+1'])

ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Cambiar la linea del eje X al 0
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))

# Cambiar la linea del eje Y al 0
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

plt.legend(loc="upper left")

# Crear una linea discontinua
x0 = 2*np.pi/3
plt.plot([x0,x0], [0,np.sin(x0)], color="blue", linewidth=1.5, linestyle="--")
plt.scatter([x0, ], [np.sin(x0), ], 50, color="blue")
plt.annotate(r'$\sin(\frac{2\pi}{3}) = \frac{\sqrt{3}}{2}$',
            xy = (x0, np.sin(x0)), xycoords = "data",
            xytext = (+20, +40), textcoords = "offset points",
            fontsize = 16, arrowprops = dict(arrowstyle = "->", connectionstyle="arc3, rad=.2"))

plt.plot([x0,x0], [0,np.cos(x0)], color="green", linewidth=1.5, linestyle="--")
plt.scatter([x0, ], [np.cos(x0), ], 50, color="green")
plt.annotate(r'$\cos(\frac{2\pi}{3}) = -\frac{1}{2}$',
            xy = (x0, np.cos(x0)), xycoords = "data",
            xytext = (-90, -60), textcoords = "offset points",
            fontsize = 16, arrowprops = dict(arrowstyle = "->", connectionstyle="arc3, rad=.2"))

for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontsize(16)
    label.set_bbox(dict(facecolor="white", edgecolor="None", alpha = 0.6))

plt.show()
```
